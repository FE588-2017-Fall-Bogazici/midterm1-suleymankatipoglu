```python
#1

%matplotlib inline
import matplotlib.pylab as plt
import numpy as np

L=200
generated_randomwalk = [np.random.randn() for i in range(L)] 
window = L+1
mav = []
for i in range(len(generated_randomwalk)):
    s = 0
    n = 0
    for j in range(window):
        if i-j < 0:
            break;
        
        s = s + generated_randomwalk[i-j]
        n = n + 1
    mav.append(s/n)

plt.plot(generated_randomwalk)
plt.plot(mav)
plt.show()
```

```python
#2
#European Option Binomial Pricing Model

import numpy as np


S0 = 100
N = 10
K = 120
u = 2
d = 1/u
r = 0.05
p = 0.5
q = 1-p
T = 1

for i in range(N):
    Wt = np.random.binomial(1, p)
    if Wt == 0:
        ST = S0*d
        S0 = ST
    else:
        ST = S0*u
        S0 = ST
        
    C += max(0, ST - K)
    P += max(0, K - ST)
    
call_price = (C/N)*np.exp(-r*T)
put_price = (P/N)*np.exp(-r*T)

print('Call price =' , call_price)
print('Put price =' , put_price)
    
```

```python
#2
#American Option Binomial Pricing Model

import numpy as np


S0 = 100
N = 10
K = 120
u = 2
d = 1/u
r = 0.05
p = 0.5
q = 1-p
T = 1
C = 0
P = 0

calllist = []
putlist = []

for i in range(N):
    Wt = np.random.binomial(1, p)
    if Wt == 0:
        ST = S0*d
        S0 = ST
        calllist.append(ST)
    else:
        ST = S0*u
        S0 = ST
        putlist.append(ST)
        
C += max(0, max(calllist))
P += max(0, min(putlist))

    
call_price = (C/N)*np.exp(-r*T)
put_price = (P/N)*np.exp(-r*T)


print('Call price =' , call_price)
print('Put price =' , put_price)
```
